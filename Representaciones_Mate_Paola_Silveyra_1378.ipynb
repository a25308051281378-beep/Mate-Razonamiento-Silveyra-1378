#### Problema 2:
Edades de Pedro y AnaPaso 1: Análisis y ResoluciónEste problema requiere relacionar dos variables basadas en una sola referencia.Planteamiento: * Edad de Ana: $x$ (es nuestra base).Edad de Pedro: $2x$ (el doble que la de Ana).Ecuación: La suma de ambas edades es 36.$$x + 2x = 36$$Simplificación: Sumamos los términos con $x$.$$3x = 36$$Despeje: El 3 que multiplica a la $x$ pasa dividiendo al 36.$$x = \frac{36}{3}$$Solución: * $x = 12$ (Edad de Ana)$2x = 2(12) = 24$ (Edad de Pedro)

```python
# Problema 2: Edades de Pedro y Ana
suma_edades = 36

# La ecuación es x + 2x = 36 -> 3x = 36
# Por lo tanto, x = suma_edades / 3
edad_ana = suma_edades / 3
edad_pedro = edad_ana * 2

print(f"La edad de Ana es: {int(edad_ana)} años")
print(f"La edad de Pedro es: {int(edad_pedro)} años")
```
